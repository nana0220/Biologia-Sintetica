```python
#Declaración de variables
%pylab inline
import time


#todo en milisegundos-->1000 ms es 1 seg, 60 seg 1 min
kr=1.0/60000.0 # producción de mRNA por minuto
gr=1.0/300000.0 #destrucción de mRNA por minuto (gamma r)
cells=10
t=60*60*1000
#3*60*60*1000 #tiempo de generación en minutos
kp=1.0/1000.0
gp=1.0/180000.0
ini=time.time()
```

$Código$ $para$ $simulación$

```python
#proceso

mRNA=zeros((t,cells))
prot=zeros((t,cells))
for j in range(cells):
    for i in range(t):
        a=random.rand(1)[0]
        b=random.rand(1)[0]
        c=random.rand(1)[0]
        d=random.rand(1)[0]
        if(a<kr and b>(gr*mRNA[i-1,j])): #creacion rna
            mRNA[i,j]=mRNA[i-1,j]+1
        elif(a>kr and b<(gr*mRNA[i-1,j])): #destruccion mrna
            mRNA[i,j]=mRNA[i-1,j]-1
        else:
            mRNA[i,j]=mRNA[i-1,j]
        if(c<(kp*mRNA[i-1,j]) and d>(gp*prot[i-1,j])):#creacion prot
            prot[i,j]=prot[i-1,j]+1
        elif(c>(kp*mRNA[i-1,j]) and d<(gp*prot[i-1,j])): #destruccion prot
            prot[i,j]=prot[i-1,j]-1
        else:
            prot[i,j]=prot[i-1,j]
            
plot(range(t),mRNA)
title('mRNA vs tiempo')
xlabel('tiempo')
ylabel('mRNA')
print (time.time()-ini)/60
           
```

```python
plot(range(t),prot)
title('Proteínas vs tiempo')
xlabel('tiempo')
ylabel('proteinas')
```

```python
rna_prom=zeros(t)
prot_prom=zeros(t)
rna_std=zeros(t)
prot_std=zeros(t)

for u in range(t):
    rna_prom[u]=mean(mRNA[u,:])
    
for k in range(t):
    prot_prom[k]=mean(prot[k,:])
    
for e in range(t):
    rna_std[e]=std(mRNA[e,:])
    
for w in range(t):
    prot_std[w]=std(prot[w,:])
    
    
plot(range(t),rna_prom)
title('rna promedio vs tiempo')
xlabel('tiempo')
ylabel('rna')
plot(range(t),prot_prom)
title('Proteínas promedio vs tiempo')
xlabel('tiempo')
ylabel('proteinas')
```

$Ruido$ $numérico$ $de$ $RNA$ $y$ $proteínas:$

```python
noise_rna=mean(rna_prom)/mean(rna_std)
noise_prot=mean(prot_prom)/mean(prot_prom)
print noise_rna,noise_prot
```

$Feedback$ $Negativo$

```python
#Feedback negativo--> ahora kr/1+(p/k)^2 donde K=1
K=2000
inic=time.time()
for j in range(cells):
    for i in range(t):    
        a=random.rand(1)[0]
        b=random.rand(1)[0]
        c=random.rand(1)[0]
        d=random.rand(1)[0]
        if(a<(kr/1+(prot[i,j]/K)**2) and b>(gr*mRNA[i-1,j])): #creacion rna
            mRNA[i,j]=mRNA[i-1,j]+1
        elif(a>(kr/1+(prot[i,j]/K)**2) and b<(gr*mRNA[i-1,j])): #destruccion mrna
            mRNA[i,j]=mRNA[i-1,j]-1
        else:
            mRNA[i,j]=mRNA[i-1,j]
        if(c<(kp*mRNA[i-1,j]) and d>(gp*prot[i-1,j])):#creacion prot
            prot[i,j]=prot[i-1,j]+1
        elif(c>(kp*mRNA[i-1,j]) and d<(gp*prot[i-1,j])): #destruccion prot
            prot[i,j]=prot[i-1,j]-1
        else:
            prot[i,j]=prot[i-1,j]
            
plot(range(t),mRNA)
title('rna vs tiempo con feedback negativo')
xlabel('tiempo')
ylabel('rna')
plot(range(t),prot)
title('Proteínas vs tiempo con feedback negativo')
xlabel('tiempo')
ylabel('proteinas')
print time.time()-inic/60
```
