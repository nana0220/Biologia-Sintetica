```python
%pylab inline
import time
kr=1.0
gr=1.0/5.0
kp=60.0
gp=1.0/30.0
cells=10
n=60000
K=2000
rna=zeros((n,cells))
prot=zeros((n,cells))
T=zeros((n,cells))
initial=time.time()
```

```python
for i in range(n-1):
    for j in range(cells):
        a=random.rand(1)[0]
        s=sum([kr,kp*rna[i,j],gr*rna[i,j],gp*prot[i,j]])
        t=(1/s)*log(1/a)
        b=random.rand(1)[0]
        T[i+1]=T[i]+t
        if(b<(kr/s)):
            rna[i+1,j]=rna[i,j]+1
            prot[i+1,j]=prot[i,j]
        elif(b>(kr/s) and b<((kr+kp*rna[i,j])/s)):
            rna[i+1,j]=rna[i,j]
            prot[i+1,j]=prot[i,j]+1
        elif(b>((kr+kp*rna[i,j])/s) and b<((kr+kp*rna[i,j]+gr*rna[i,j])/s)):
            rna[i+1,j]=rna[i,j]-1
            prot[i+1,j]=prot[i,j]
        elif(b>((kr+kp*rna[i,j]+gr*rna[i,j])/s)):
            rna[i+1,j]=rna[i,j]
            prot[i+1,j]=prot[i,j]-1
        

plot(T,prot)
title('Proteínas vs tiempo')
xlabel('tiempo')
ylabel('proteinas')
final=time.time()
lenght=final-initial
print lenght/60
```

```python
#normalizacion
interval=100.0*60.0*10.0
t_norm=linspace(0,120.0,interval)
rna_norm=zeros((size(t_norm),cells))
prot_norm=zeros((size(t_norm),cells))

j=0

for i in range(cells):
    for k in t_norm:
        while((T[j,i]*600.0)<t_norm[k]):
            j=j+1
            rna_norm[k,i]=rna[j,i]
            prot_norm[k,i]=prot[j,i]
            

t_norm
rna_norm
            
```

$Feedback$

```python
for i in range(n-1):
    for j in range(cells):
        a=random.rand(1)[0]
        s=sum([kr/(1+(prot[i,j]/K)**2),kp*rna[i,j],gr*rna[i,j],gp*prot[i,j]])
        t=(1/s)*log(1/a)
        b=random.rand(1)[0]
        T[i+1]=T[i]+t
        if(b<((kr/(1+(prot[i,j]/K)**2))/s)):
            rna[i+1,j]=rna[i,j]+1
            prot[i+1,j]=prot[i,j]
        elif(b>((kr/(1+(prot[i,j]/K)**2))/s) and b<((kr+kp*rna[i,j])/s)):
            rna[i+1,j]=rna[i,j]
            prot[i+1,j]=prot[i,j]+1
        elif(b>(((kr/(1+(prot[i,j]/K)**2))+kp*rna[i,j])/s) and b<((kr+kp*rna[i,j]+gr*rna[i,j])/s)):
            rna[i+1,j]=rna[i,j]-1
            prot[i+1,j]=prot[i,j]
        elif(b>(((kr/(1+(prot[i,j]/K)**2))+kp*rna[i,j]+gr*rna[i,j])/s)):
            rna[i+1,j]=rna[i,j]
            prot[i+1,j]=prot[i,j]-1
        

plot(T,prot)
title('Proteínas vs tiempo con feedback')
xlabel('tiempo')
ylabel('proteinas')
```
